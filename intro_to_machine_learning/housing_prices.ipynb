First import Pandas and reach the data file (in this case 'melb_data.csv')

```python
# Import Pandas
import pandas as pd

MELB_DATASET_PATH = 'melb_data.csv'

# Read the dataset
melbourne_data = pd.read_csv(MELB_DATASET_PATH)

# Display the DataFrame metadata
melbourne_data.describe()
```

Next drop na values and select the prediction target using dot notation and assign it to the variable y

```python
# Drop na values
melbourne_data = melbourne_data.dropna(axis=0)

# Capture prediction target
y = melbourne_data.Price
```

Select the features the model will take as inputs and assign them to the variable X

```python
# Create a list of the table columns to use as features
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

# Select the features from the dataset
X = melbourne_data[melbourne_features]
```

Dont forget to examine each data set to make sure nothing wierd is going on!

```python
# Display the features DataFrame metadata
X.describe()
```

```python
# Display the first 5 rows of the features DataFrame
X.head()
```

Next import the decision tree regression model from scikit-learn 

```python
# import scikit-learn
from sklearn.tree import DecisionTreeRegressor

# Define the model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit the model to the training data
melbourne_model.fit(X, y)
```

Next we're going to make some predicitons for the first 5 houses in the data set to see how good our model is.

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

This gives us a model that can make predicitons very well for this training set. Next we need to evaluate how well the model does at predicting new inputs.
To do this the first and most basic metric we can use is the Mean Absolute Error (MAE).

```python
from sklearn.metrics import mean_absolute_error

# get the predictions from the model
predicted_home_prices = melbourne_model.predict(X)

# calculate the mean absolute error of the prediction
mean_absolute_error(y, predicted_home_prices)
```

This is an "in sample" score meaning we are evaluating the error on the training data which is a bad practice and can lead to over fitting.
To combat this overfitting we can use the train_test_split function from sklearn.

```python
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

# Define the model.
melbourne_model = DecisionTreeRegressor()

# Fit the model to the training data.
melbourne_model.fit(train_X, train_y)

# get the predicted prices on validation data
val_predicitons = melbourne_model.predict(val_X)

#get the mean absolute error of the prediction
print(mean_absolute_error(val_y, val_predicitons))
```

Underfitting and Overfitting:

Underfitting happens when the model doesn't capture important distinctions and patterns in a dataset and therefore generates poor predictions on both the training and validation data.

Overfitting is when the model predicts the training set almost perfectly but performs poorly on the validation data.

A ultility function can be used to iterate over multiple different model configurations to find what fits the data best

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    val_predictions = model.predict(val_X)
    return mean_absolute_error(val_y, val_predictions)
```

Then we can loop over different values for the max_leaf_nodes to see what gets us the best predicitons.

```python
for max_leaf_nodes in [5, 50, 100, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" % (max_leaf_nodes, my_mae))
```

The decision tree model can only take us so far, there is a more accurate model known as the "random forest" which uses many trees and makes a prediction using the average of all the trees.

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

The kwarg n_estimators can be used in RandomForestRegressor to choose the number of trees
